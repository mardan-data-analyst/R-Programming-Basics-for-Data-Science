### **3.9 Debugging**

**Producing an error**

```r
# "a" + 10
# Error in "a" + 10: non-numeric argument to binary operator
# Traceback:
```

**Error halts execution**

```r
# for(i in 1:3) {
    # print(i + "a")
# }

# Error in i + "a": non-numeric argument to binary operator
# Traceback:

# 1. print(i + "a")
```

**Error catching with tryCatch**

```r
# normal execution
tryCatch(10 + 10)
```

```r
# error
# tryCatch("a" + 10)

# Error in "a" + 10: non-numeric argument to binary operator
# Traceback:

# 1. tryCatch("a" + 10)
# 2. tryCatchList(expr, classes, parentenv, handlers)
```

```r
tryCatch(10 + "a", error = function(e)
    print("Something went wrong."))
```

```r
tryCatch(10 + 10, error = function(e)
    print("Something went wrong."))
```

```r
tryCatch(10 + "a", error = function(e)
    return("10a"))
```

```r
tryCatch(
    for(i in 1:3) {
        print(i + "a")
    }, error = function(e) print("Found error."))
```

**Warning hanndling**

```r
# as.integer("A")

# Warning message in eval(expr, envir, enclos):
# “NAs introduced by coercion”
# <NA>
```

```r
tryCatch(
    as.integer("A"),
    warning = function(e)
        print("Warning!")
)
```
